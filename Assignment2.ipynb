**Labeling Data**

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
X
y
```

**Train Test Split**

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,  train_size =0.7, random_state = 100)
```

**Applying Standard Scalar: To standardize data**

```python
from sklearn.preprocessing import StandardScaler
#from sklearn.preprocessing import MinMaxScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test
```

**Applying Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression
classification=LogisticRegression(solver='lbfgs', max_iter=5000)
classification.fit(X_train_scaled,y_train)
```

```python
y_pred=classification.predict(X_test)
```

```python
classification.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
#y_pred = random.predict(X_test)
print(classification_report(y_test,y_pred))
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

#Setup arrays to store training and test accuracies
neighbors = np.arange(1,7)
train_accuracy =np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)
    
    #Fit the model
    knn.fit(X_train_scaled, y_train)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train_scaled,y_test)
    
    #Compute accuracy on the test set
    test_accuracy[i] = knn.score(X_train_scaled, y_test) 
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(splitter="best", random_state = 0)
reg.fit(X_train_scaled,y_train)
reg.predict(X_test)
score = reg.score(X_test, y_test)
print(score)
```

```python
from sklearn.linear_model import Perceptron
model = Perceptron()
model.fit(X_train_scaled,y_train)
prediction=model.predict(X_test)
print('Predicted: %d' % prediction)
```

```python
knn = KNeighborsClassifier(n_neighbors=5) #Using Eucledian distance
knn.fit(X_train_scaled,y_train)
Accuracy=knn.score(X_test,y_test)
print('KNN Accuracy:',Accuracy)
```
